### Basic working of Google Palm LLM in LangChain

```python
from langchain.llms import GooglePalm

llm = GooglePalm(google_api_key=api_key, temperature=0.1)
```

```python
poem = llm("Write a 4 line poem of my love for dosa")

print(poem)
```

```python
essay = llm("write email requesting refund for electronic item")
print(essay)
```

### Now let's load data from Codebasics FAQ csv file

```python
from langchain.document_loaders.csv_loader import CSVLoader

loader = CSVLoader(file_path='/content/sample_data/codebasics_faqs3.csv', source_column='prompt')

# Store the loaded data in the 'data' variable
data = loader.load()
data
```

### Hugging Face Embeddings

```python
from langchain.embeddings import HuggingFaceInstructEmbeddings

# Initialize instructor embeddings using the Hugging Face model
instructor_embeddings = HuggingFaceInstructEmbeddings(model_name="hkunlp/instructor-large")

e = instructor_embeddings.embed_query("What is your refund policy?")
```

```python
len(e)
```

```python
e[0:5]
```

### Vector store using FAISS

```python
from langchain.vectorstores import FAISS

# Create a FAISS instance for vector database from 'data'
vectordb = FAISS.from_documents(documents=data,
                                 embedding=instructor_embeddings)

# Create a retriever for querying the vector database
retriever = vectordb.as_retriever(score_threshold = 0.7)
```

```python
rdocs = retriever.get_relevant_documents("why should I trust on this course?")
rdocs
```

### Create RetrievalQA chain  🚀

---



```python
from langchain.chains import RetrievalQA

chain = RetrievalQA.from_chain_type(llm=llm,
                            chain_type="stuff",
                            retriever=retriever,
                            input_key="query",
                            return_source_documents=True)
chain("how many people has subscribed so far ?")
```

### Create RetrievalQA chain along with prompt template 🚀

---



```python
from langchain.prompts import PromptTemplate

prompt_template = """Given the following context and a question, generate an answer based on this context only.
In the answer try to provide as much text as possible from "response" section in the source document context without making much changes.
If the answer is not found in the context, kindly state "I don't know." Don't try to make up an answer.

CONTEXT: {context}

QUESTION: {question}"""


PROMPT = PromptTemplate(
    template=prompt_template, input_variables=["context", "question"]
)
chain_type_kwargs = {"prompt": PROMPT}


from langchain.chains import RetrievalQA

chain = RetrievalQA.from_chain_type(llm=llm,
                            chain_type="stuff",
                            retriever=retriever,
                            input_key="query",
                            return_source_documents=True,
                            chain_type_kwargs=chain_type_kwargs)
```
